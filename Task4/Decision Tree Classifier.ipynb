```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('Iris.csv')
data.head()
```

```python
data.drop(['Id'],axis=1,inplace=True)
```

```python
data.head()
```

```python
X=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# create Decision Tree Classifier
treeclf = DecisionTreeClassifier()
treeclf.fit(X_train, y_train)
```

### Visualize Decision Tree Classifier

```python
# Import necessary libraries for graph viz
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
```

```python
# Visualize the graph

export_graphviz(treeclf, out_file="decision_tree.odt", feature_names=X.columns,  filled=True, rounded=True,special_characters=True)
graph = pydotplus.graph_from_dot_file("decision_tree.odt")  
Image(graph.create_png())
```

### Making predictions for the testing data

```python
y_pred=treeclf.predict(X_test)
y_pred
```

```python
print(accuracy_score(y_test, y_pred))
```
